**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
#create function to calculate ratio of house prices

def create_price_ratio(data_filter, start_year='1998', end_year='2018'):
    
    def calculate_annual_average(monthly_data_for_year):
        return monthly_data_for_year.mean()
    
    start_year_prices = data_filter[data_filter['Year'] == start_year]['Average_Price']
    end_year_prices = data_filter[data_filter['Year'] == end_year]['Average_Price']
    
    start_year_mean = calculate_annual_average(start_year_prices)
    end_year_mean = calculate_annual_average(end_year_prices)
    
    return end_year_mean / start_year_mean
        
```

```python
#calculate the price change ratio for each borrow for 1998 to 2018

list_of_ratio_dicts = []

for borough in property_data['Borough_Name'].unique():
    data_filter = property_data[property_data['Borough_Name'] == borough]
    ratio = create_price_ratio(data_filter)
    ratio_dict = {'Borough_Name':borough, 'Ratio':ratio}
    list_of_ratio_dicts.append(ratio_dict)

df_borough_ratio = pd.DataFrame(list_of_ratio_dicts)
```

```python
#sort the borough ratios to determine which borough had the 
#greatest price increase between 1998 and 2018

df_borough_ratio.sort_values('Ratio', ascending=False)
```

```python
#plot the data to visualize differences in price increase between boroughs

borough_ratio_bar_plot = df_borough_ratio.sort_values('Ratio')\
    .plot(kind="bar", x='Borough_Name', y='Ratio', \
          title='Housing price change ratio for London boroughs (1998-2018)')
borough_ratio_bar_plot.set_ylabel('Ratio')
borough_ratio_bar_plot.get_legend().remove()
```
